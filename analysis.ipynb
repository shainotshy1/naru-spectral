```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = [80, 96, 115, 138, 166, 199, 239, 287, 344]
r2_spec = [0.16362354507716248, 0.8249651546835807, 0.9148912153243804, 0.9486160688746367, 0.9491028110388907, 0.9711965300768052, 0.9962698413804532, 0.9949353029739274, 0.9937501752884327]
r2_lasso = [-0.303413114970525, 0.6798347309865804, 0.6408555650014071, 0.6321196137672989, 0.6661309668826633, 0.6635155948600221, 0.6558470841658105, 0.66, 0.6674120833459011]
```

```python
N = min(len(n), len(r2_spec), len(r2_lasso))
plt.plot(n[:N], r2_spec[:N], marker='o', label='DBOOST')
plt.plot(n[:N], r2_lasso[:N], marker='o', label='Lasso')
# plt.axhline(0, color='gray', linestyle='--')
# plt.xscale('log')
plt.ylim(0)
plt.xlabel('Number of Samples')
R2 = r'$R^2$'
plt.ylabel(f'Faithfulness ({R2})')
plt.title('Cardinality Estimator Faithfulness: DMV')
plt.legend()
plt.show()
```
